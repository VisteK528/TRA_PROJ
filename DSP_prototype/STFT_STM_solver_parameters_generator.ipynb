```python
fft_size = 256
hann_window = np.hanning(fft_size)

print("const float hann_window[FFT_SIZE] = {")
for i in range(0, len(hann_window), 10):
    line = ", ".join(f"{x:.7f}" for x in hann_window[i:i+10])
    if i + 10 < len(hann_window):
        print(f"    {line},")
    else:
        print(f"    {line}")
print("};")
```

```python
fs = 8000
f_zero = 0
f_pass = 120
minimum_transmittance = 1e-3

stft_filter_mask, zero_index, pass_index = DSP_algorithms.get_highpass_stft_filter_mask(mask_length=129, fs=fs, f_pass=f_pass,
                                                                                        f_zero=f_zero,
                                                                                        minimum_transmittance=minimum_transmittance,
                                                                                        window_type='hann')
print(f"zero_index: {zero_index}")
print(f"pass_index: {pass_index}\n\n")
stft_filter_mask = stft_filter_mask[zero_index:pass_index]

print("const float stft_filter_mask[MASK_SIZE] = {" + (", ".join(f"{x:.7f}" for x in stft_filter_mask)) + "};")
```
