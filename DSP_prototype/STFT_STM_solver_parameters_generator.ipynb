```python
import numpy as np
```

```python
fft_size = 256
hann_window = np.hanning(fft_size)

print("const float hann_window[FFT_SIZE] = {")
for i in range(0, len(hann_window), 10):
    line = ", ".join(f"{x:.7f}" for x in hann_window[i:i+10])
    if i + 10 < len(hann_window):
        print(f"    {line},")
    else:
        print(f"    {line}")
print("};")
```
